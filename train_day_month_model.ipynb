```python
from day_month_data_pipeline import DayMonth_data_pipeline
from cnn_model import DayMonth_Predictor
```

```python
day_month_data = DayMonth_data_pipeline(data_dir="dataset/day_month")
train_ds, val_ds = day_month_data.get_train_val_data()
```

```python
predictor = DayMonth_Predictor()
predictor.build_cnn()
predictor.compile_model()
```

```python
print("Training started")
history = predictor.train_model(train_ds, val_ds, epochs=50)
save_path = "trained_models/day_month_model/day_month_model.h5"
predictor.model.save(save_path)
print("Model is saved successfully")
```

```python
results = predictor.model.evaluate(val_ds)
print(f"Validation Loss: {results[0]}")
print(f"Validation Accuracy: {results[1] * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt


plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
